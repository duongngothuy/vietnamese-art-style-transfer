# Research Question

How does access to mental health care–including therapists, facilities, and insurance coverage–differ between high-income and low-income neighborhoods across California?



## Background and Prior Work

Mental health care is an essential component of health care, but it is still one of the most neglected areas within the U.S. healthcare system. Millions of people in the U.S. experience many different mental health conditions every year, with those conditions ranging from anxiety disorders, depression, or PTSD, and that experience is different for everyone, especially for marginalized communities. Communities of a lower socioeconomic status are more likely to be at the higher end of barriers due to their lower income status. These barriers can range in cost, not having a provider in their community, no insurance or underinsured, and long wait times for care or treatment. Because of this, we are leaving people behind without care and without the support they need when they need it most, further perpetuating and contributing to such health disparities.

Research has repeatedly established that low-income and minority communities experience unreasonably more difficulty in accessing mental health care. As reported by the Centers for Disease Control and Prevention (CDC), adults living at or below the national poverty level are more likely to report frequent mental distress; but they are also the least likely to receive adequate treatment.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) The National Alliance for Mental Illness (NAMI) reports that many providers will not accept public insurance programs (e.g. Medicaid) that are often used by low-income populations.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Additionally, there are treatment access disparities for racial and ethnic minorities that were limited for Latinx, Asians, and Black communities, as demonstrated in a study in 2018 by Alegría et al., which highlighted notable barriers in access to treatment that are tied to distance and economy.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Even if these studies indicate widespread equality, research has mostly focused on broad inequalities or large-scale statistics. Despite those sources that indicate neighborhood-level inequalities, especially urban environments that can differ from communities to communities, there does not appear to be much research that compares access at the neighborhood level. Access to care can vary dramatically depending on income level, zoning laws, and clinic location. Our project aims to complete a purpose that compares mental health resources in low-income neighborhoods to high-income neighborhoods using publicly available data. By investigating issues including number of providers, provider acceptance of insurance, and more specific variables in the project, this analysis hopes to contribute to a more practical understanding of access to mental health resources at a local level.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

Our hypothesis is that lower-income neighborhoods in California will have significantly less access to mental health resources compared to higher-income neighborhoods. We expect that there will be fewer licensed mental health providers per capita, lower availability of health insurance, and a correlation between provider distribution and neighborhood income levels.
```python
import pandas as pd
import numpy as np
import requests as re
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('talk')
```

```python
ca_zipcodes = pd.read_csv('data/cazipcodes.csv')
ca_zipcodes = ca_zipcodes[['zip', 'city', 'county']]
ca_zipcodes['zip'] = ca_zipcodes['zip'].astype(int)
ca_zipcodes
```

## Dataset #2: 2023 American Community Survey 5-Year Estimate

```python
# collect population data and median household income indexed by zip codes from US 2023 ACS 5-year data API

data = re.get(f'https://api.census.gov/data/2023/acs/acs5?get=B01003_001E,B19013_001E&for=zip%20code%20tabulation%20area:*&key=3c64d895f18a7dc61b53f1482d7c9d8b12ea4fe7')
data_json = data.json()
```

```python
# format data into dataframe

def df_clean_up(df):
    df = df.rename(columns={'B01003_001E': 'population_total',
                            'B19013_001E': 'median_hh_income_total',
                            'zip code tabulation area': 'zip_code'})
    
    df['zip_code'] = df['zip_code'].astype(int) #convert zip codes to integers
    df['population_total'] = df['population_total'].astype(int).replace(0, np.nan) # replaced populations of 0 with null
    df['median_hh_income_total'] = df['median_hh_income_total'].astype(int).replace(-666666666.0, np.nan) # replaced missing incomes with null

    return df

df = pd.DataFrame(data_json[1:], columns=data_json[0]).pipe(df_clean_up)
df
```

```python
# merge california zip codes with US data

zip_data = df.merge(ca_zipcodes, left_on='zip_code', right_on='zip', how='inner')
zip_data = zip_data.drop(columns=['zip'])
zip_data = zip_data[zip_data['median_hh_income_total'].isna() == False] #dropped zip codes without median household income info (209 values)
zip_data = zip_data.set_index('zip_code')
zip_data
```

## Dataset #4: Therapist by Zip Code in California

```python
# Read the file
df = pd.read_csv("data/Psychology_Data00.xls", sep='\t')
```

```python
# Filter rows to keep only License Type containing Psychologist
psychologists = df[df['License Type'].str.contains('Psychologist', case=False, na=False)]

# Filter to only keep data from those in California 
psychologists_ca = psychologists[psychologists['State'] == 'CA']

# Keep Zip and County columns
df_selected = psychologists_ca[['Zip', 'County']].copy()

# Replace NaN values with string '0'
df_selected = df_selected.fillna('0')

# Keep only the first 5 digits of Zip code as string
df_selected['Zip'] = df_selected['Zip'].astype(str).str[:5]

# Convert column names to lowercase
df_selected.columns = df_selected.columns.str.lower()

# Group by both zip and county, and count number of psychologists
psychologists_by_zip = (
    df_selected.groupby(['zip', 'county'])
    .size()
    .reset_index(name='psychologist_count')
)

# Sort by count descending
psychologists_by_zip = psychologists_by_zip.sort_values(by='psychologist_count', ascending=False)

# Preview
print(psychologists_by_zip)

# Take top 20 rows and make a copy to avoid SettingWithCopyWarning
top_20 = psychologists_by_zip.head(20).copy()

# Create combined label for zip and county
top_20['zip_county'] = top_20['zip'] + ', ' + top_20['county']

# Plot
plt.figure(figsize=(12,8))
plt.barh(top_20['zip_county'], top_20['psychologist_count'], color='skyblue')
plt.xlabel('Number of Psychologists')
plt.ylabel('Zip Codes, County')
plt.title('Top 20 California ZIP Code + County Areas by Number of Psychologists')
plt.gca().invert_yaxis()  # largest bars on top
plt.tight_layout()
plt.show()
```

## Dataset #5: Covered California 2025 Products by Zip Code

```python
#Read Covered California health insurance by zip code csv file into dataframe
path = 'data/CAProductByZip.csv'
data = pd.read_csv(path)
data
```

```python
#Remove duplicate zip codes from county's that share zip codes
data_unique = data.drop_duplicates(subset="Zip Code", keep="first")

#Change both zip code columns to have the same format
data_unique['Zip Code'] = data_unique['Zip Code'].astype(int)
ca_zipcodes['zip'] = ca_zipcodes['zip'].astype(int)

#rename zip column in ca_zipcodes to match data
data2_renamed = ca_zipcodes.rename(columns={'zip': 'Zip Code'})

#merge on inner "Zip Code" to keep matching zip codes
merge_data = pd.merge(data_unique, data2_renamed, on='Zip Code', how='inner')

merge_data.shape
merge_data
```

```python
#List of product columns
product_columns = ['Aetna HMO', 'Anthem EPO', 'Anthem HMO', 'Blue Shield HMO', 
                   'Blue Shield PPO', 'Chinese Community HMO', 'Health Net HMO', 
                   'Health Net PPO', 'IEHP HMO', 'Kaiser HMO', 'LA Care HMO', 
                   'Molina HMO', 'Sharp HMO 1 (Copay)', 'Sharp HMO 2 (Coin)']

#When xlsx file is convert to csv file, the 'X's were converted to random numbers 1-9, so they must be replaced
#Replace all non-NaN values in the product columns with 1
merge_data[product_columns] = merge_data[product_columns].notna().astype(int)

#get sum of health products offered per zip code
df_total_products = merge_data.groupby('Zip Code')[product_columns].sum()

#add a new column for the total products offered per Zip Code
df_total_products['Total Products'] = df_total_products.sum(axis=1)

#add a new DataFrame that is for zip code and total products offered per zip code
df_total_products = df_total_products[['Total Products']].reset_index()

print(df_total_products)
```

```python
#histogram of number of zip codes across number of products offered, kde shows frequency of data
plt.figure(figsize=(10, 6))
sns.histplot(df_total_products['Total Products'], kde=True, bins=30, color='purple')
#sns.histplot(df_total_products['Total Products'], bins=30)

#Labels
plt.xlabel('Total Healthcare Products')
plt.ylabel('Number of Zip Codes')
plt.title('Distribution of Healthcare Products Across Zip Codes')

#show plot
plt.tight_layout()
plt.show()
```

### Section 2 of EDA: Merged Income Limits and Insurance Product Availability Dataset for California Counties (2022–2023)

#### What We Learned From This EDA from Dataset #2 and Dataset #4

Through EDA, we examined the relationship between county-level income limits and the availability of health insurance products in California. By combining datasets on Area Median Income (AMI) and the number of insurance plans offered by zip code, we visualized how economic conditions connect with healthcare access.


Our scatter plot showed that counties with higher AMI generally offer more health insurance options. A boxplot comparing high-income and low-income counties added on to this finding, showing a significant gap in average product availability. These patterns suggest that residents in lower-income areas may experience less access to mental health-related insurance resources. These two datasets help guide our broader analysis of disparities in mental health care across the state.


#### How These Datasets Connect


The income limits dataset provides economic context by reporting Area Median Income (AMI) and standardized income brackets (e.g., Acutely Low Income, Very Low Income) for each California county. This allowed us to identify which counties fall into higher or lower income categories based on official thresholds.


The second dataset tracks how many health insurance products are available in each zip code, which we aggregated to the county level. This gives us a proxy for how much healthcare choice, and by extension, mental health access is available in each region.


By merging these two datasets using county as a common factor, we were able to analyze whether income level influences access to health insurance plans. This merged dataset directly supports our research question about whether lower-income neighborhoods face reduced access to mental health care.


```python
#STEP 1 & 2:
import pandas as pd


# Load the datasets
df_insurance = pd.read_csv("data/CAProductByZip.csv")
df_zip_ref = pd.read_csv("data/cazipcodes.csv")


# Drop duplicate zip codes and make an explicit copy
df_insurance_unique = df_insurance.drop_duplicates(subset="Zip Code", keep="first").copy()


# Standardize zip code formats to 5 digits
df_insurance_unique['Zip Code'] = df_insurance_unique['Zip Code'].astype(str).str.zfill(5)
df_zip_ref['zip'] = df_zip_ref['zip'].astype(str).str.zfill(5)


# Rename 'zip' column to match for merging
df_zip_ref_renamed = df_zip_ref.rename(columns={'zip': 'Zip Code'})


# Merge the insurance and zip reference datasets on Zip Code
df_merged = pd.merge(df_insurance_unique, df_zip_ref_renamed, on='Zip Code', how='inner')


# Preview
print(f"Merged dataset shape: {df_merged.shape}")
df_merged.head()
```

```python
# STEP 3: Calculate Total Health Insurance Products per Zip Code


# List of columns that represent available health insurance plans
product_columns = [
    'Aetna HMO', 'Anthem EPO', 'Anthem HMO', 'Blue Shield HMO',
    'Blue Shield PPO', 'Chinese Community HMO', 'Health Net HMO',
    'Health Net PPO', 'IEHP HMO', 'Kaiser HMO', 'LA Care HMO',
    'Molina HMO', 'Sharp HMO 1 (Copay)', 'Sharp HMO 2 (Coin)',
    'Valley HMO', 'Western Health HMO'
]


# Convert all non-null entries in product columns to 1, nulls to 0
# This creates a binary indicator for whether a product is offered
df_merged[product_columns] = df_merged[product_columns].notna().astype(int)


# Calculate the total number of products offered in each zip code
df_merged['Total Products'] = df_merged[product_columns].sum(axis=1)


# Create a cleaned DataFrame with just the info we need
df_total_products = df_merged[['Zip Code', 'county', 'Total Products']]


# Preview the result
df_total_products.head()
```

```python
# STEP 4: Group by County to calculate average number of products per zip code


# Group zip-level product totals to the county level
df_products_by_county = df_total_products.groupby('county')['Total Products'].mean().reset_index()


# Rename column for clarity
df_products_by_county = df_products_by_county.rename(columns={'Total Products': 'Avg Products Per Zip'})


# Preview the result
df_products_by_county.head()
```

```python
#STEP 5: Combine two datasets and cleaning needed 
# Reload combined income data (if needed)
df_combined = pd.concat([df_2022, df_2023], ignore_index=True)


# Rename 'County' column for merging
df_combined_renamed = df_combined.rename(columns={'County': 'county'})


# Fix mismatch: Add " County" to county names to match insurance dataset
df_combined_renamed['county'] = df_combined_renamed['county'].astype(str) + " County"


# Merge on county
df_merged_final = pd.merge(df_combined_renamed, df_products_by_county, on='county', how='left')


# Preview
df_merged_final[['county', 'Year', 'AMI', 'Avg Products Per Zip']].head()
```

```python
# STEP 6: Scatterplot 


# Separate by year
df_2022 = df_merged_final[df_merged_final['Year'] == 2022]
df_2023 = df_merged_final[df_merged_final['Year'] == 2023]


# Create scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(df_2022['AMI'], df_2022['Avg Products Per Zip'], color='blue', alpha=0.6, label='2022')
plt.scatter(df_2023['AMI'], df_2023['Avg Products Per Zip'], color='red', alpha=0.6, label='2023')


# Add labels
plt.title("AMI vs. Avg Insurance Products per Zip Code")
plt.xlabel("Area Median Income (AMI)")
plt.ylabel("Average Products per Zip Code")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create a new column to categorize counties by income level
ami_median = df_merged_final['AMI'].median()
df_merged_final['Income Group'] = df_merged_final['AMI'].apply(
    lambda x: 'Low Income' if x < ami_median else 'High Income'
)


# Create boxplot: Average insurance products per zip code by income group
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))


low_income = df_merged_final[df_merged_final['Income Group'] == 'Low Income']['Avg Products Per Zip']
high_income = df_merged_final[df_merged_final['Income Group'] == 'High Income']['Avg Products Per Zip']


plt.boxplot([low_income.dropna(), high_income.dropna()], labels=['Low Income', 'High Income'])
plt.title('Avg Insurance Products per Zip by Income Group')
plt.ylabel('Average Products per Zip Code')
plt.grid(True)
plt.tight_layout()
plt.show()
```

This boxplot compares average health insurance product availability between low-income and high-income counties. The results clearly show that high-income counties tend to offer a wider variety of insurance options, with a higher median and broader distribution. In contrast, low-income counties are experiencing lower availability of health insurance options.
